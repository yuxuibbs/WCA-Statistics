## Which events are correlated with each other?

use everyone's best average because people can get really lucky with singles (in competition, I have a 5x5 personal best single of 1:33.53 and a personal best average of 1:50.94 and I average about 2 min at home)

use singles for the bindfolded events (3x3x3 blindfolded, 4x4x4 blindfolded, 5x5x5 blindfolded, and 3x3x3 multiple blindfolded) because it is very difficult to get 3 successes in a row and only 3x3x3 blindfolded uses mean (not many people have them either) and everything else is a best of 1 event

```python
# get events list
eventsList = list(events['id'])
# remove events that are no longer official events
eventsList.remove('333mbo')
eventsList.remove('magic')
eventsList.remove('mmagic')
eventsList
```

```python
# remove 3bld average and replace it with 3bld single and add 4bld and 5bld singles because 4bld and 5bld are best of 1 events
personalBests = ranksAverage[~(ranksAverage['eventId'] == '333bf')].append([ranksSingle[ranksSingle['eventId'] == '333bf'], 
                                                                           ranksSingle[ranksSingle['eventId'] == '444bf'], 
                                                                           ranksSingle[ranksSingle['eventId'] == '555bf'],
                                                                           ranksSingle[ranksSingle['eventId'] == '333mbf']])
# remove rankings because they are not relevent in this context
personalBests = personalBests.drop(['worldRank', 'continentRank', 'countryRank'], axis=1)
personalBests
```

```python
# create correlation matrix to calculate correlation between 2 events
corrMatrix = pd.DataFrame(index=eventsList, columns=eventsList, dtype=float)

# iterate through every event combination
for event1 in eventsList:
    for event2 in eventsList:
        # slight optimization
        if not event1 == event2:
            # calculate correlation between event1 and event2
            tempMatrix = personalBests[personalBests['eventId'] == event1].merge(personalBests[personalBests['eventId'] == event2], on='personId', suffixes=[event1, event2]).corr()
            # save result into corrMatrix
            corrMatrix.at[event1, event2] = tempMatrix['best' + event1]['best' + event2]
        else:
            corrMatrix.at[event1, event2] = 1.0
# view corrMatrix
corrMatrix
```

```python
# create heat map
sns.heatmap(corrMatrix, yticklabels=True, xticklabels=True, square=True)
```

## Which cities or regions in the United States and in the world are the most competitive?

Use tableau for the maps after preprocessing

preprocessing will be only considering competitions that are:
 * not world/regional championships because they almost always have extremely low averages
 * final rounds for the 3x3x3 event 
 
and the measure that will be used is the average of the top 3 results (average of the "podium" results) for each competition

```python
# process the data so it is easier to use
# only get competitions that are not world/regional championships
result = competitions[~competitions['id'].isin(championships['competition_id'])]

# code from office hours on 12/5
result = pd.concat([result, pd.DataFrame(result['cityName'].str.split(', ', expand=True))], axis=1)

# only get 3x3 final podium results
regionalData = results[(results['pos'] < 4) & (results['eventId'] == '333') & (results['roundTypeId'] == 'f')].groupby('competitionId').mean()
# take the average of the podium results
regionalData['competitionId'] = regionalData.index
# merge the tables so longitude/latitude is attached to competition results
output = result.merge(regionalData, left_on='id', right_on='competitionId')
# save as csv for use in Tableau
output.to_csv('competitive.csv')
output
```

# How is the community growing and how has the gender gap changed?

```python
# make data frame with just year and gender
# first 4 characters of a WCA ID is the year of the competitor's first competition
df = pd.concat([persons['id'].astype(str).str[:4], persons['gender']], axis=1)
# split into male and female
males = df[df['gender'] == 'm']
females = df[df['gender'] == 'f']
# number of males and females per year
numMalesPerYear = males.groupby('id').count()
numFemalesPerYear = females.groupby('id').count()
```

```python
# number of new people that joined the WCA every year
df.groupby('id').count()
```

```python
# total number of people in WCA every year
df.groupby('id').count().cumsum()
```

```python
# number of new males joining WCA per year
numMalesPerYear
```

```python
# total number of males in WCA every year
numMalesPerYear.cumsum()
```

```python
# graph of the total number of people in the WCA
(df.groupby('id').count().cumsum()).plot(kind='line', legend=False)
```

```python
# graph of the number of new people in the WCA per year
(df.groupby('id').count()).plot(kind='line', legend=False)
```

Had a lot of trouble overwriting the x axis and legend

```python
# graph of the total number of males and females per year
# code from office hours with Kai on 12/8
fig, ax = plt.subplots()
ax.set_ylabel('number')
ax.set_xlabel('year')
totalMalesPerYear = numMalesPerYear.cumsum()
totalFemalesPerYear = numFemalesPerYear.cumsum()
totalMalesPerYear.plot(kind='line', ax=ax, color='blue')
totalFemalesPerYear.plot(kind='line', ax=ax, color='red')
plt.show()
```

```python
# graph of the number of new males and females per year
fig, ax = plt.subplots()
ax.set_ylabel('number')
ax.set_xlabel('year')
numMalesPerYear.plot(kind='line', ax=ax, color='blue', legend=False)
numFemalesPerYear.plot(kind='line', ax=ax, color='red', legend=False)
plt.show()
```

```python
# percentage of males and females that are in the WCA every year
percentMalesPerYear = totalMalesPerYear / df.groupby('id').count().cumsum()
percentFemalesPerYear = totalFemalesPerYear / df.groupby('id').count().cumsum()
fig, ax = plt.subplots()
ax.set_ylabel('number')
ax.set_xlabel('year')
percentMalesPerYear.plot(kind='line', ax=ax, color='blue')
percentFemalesPerYear.plot(kind='line', ax=ax, color='red')
plt.show()
```
